# Excercitiul 1

Acesta este un program care ia ca input un fisier .csv cu o lista de studenti din grupa B2 si are ca output un numar stabilit alesi random, fara repetitie, de studenti. Pentru acest lucru apelez functia sample din modulul random. 

```python
import pandas as pd
import random 

def select_students_from_csv(file_path,number_selected):
    
    try:
         data=pd.read_csv(file_path)
         
         
         if 'Nume' not in data.columns:
             print ("Column 'Nume' was not found")
             return 
         
         students=data['Nume'].tolist()
         
         if number_selected>len(students):
             print("The selected number exceeds the number of avalaible students.")
             return 
         
         selected_students=random.sample(students,number_selected)
         
         print (f"Selected students ({number_selected}):")
         for student in selected_students:
             print(student)
             
    except FileNotFoundError:
        print(f"File '{file_path}' was not found.")
        
    except Exception as e:
        print(f"An error has occurred: {e}")
        
        
file_path="D:\Downloads\studenti.csv"
number_selected=3
select_students_from_csv(file_path,number_selected)
```

```python
import random 

def game(heads=0.5):
    n=0    # nr pasi pana la primul succes
    suma=0  # ce castiga primul jucator 
    
    while True:
        n+=1
        # ce pica?
        if random.random()<heads: #uuu am stema
            # trec la zar
            zar=random.randint(1,6)
            print (f"rezultatul zarului este: {zar}")
            suma+=zar-3 
            # cazul cu suma negativa e la fel daca scot in fata un minus pt ca jucatorul pierde opusul sumei practic 
            break 
        else:  #ban ban ban
            #print ("aici sunt")
            suma-=0.5   #pierde 0.5 
            
    return (n, suma)

n,suma=game()
print(f"Numarul de pasi este: {n}, Suma totala a primului jucator este: {suma} $")
            
    
```

```python
import matplotlib.pyplot as plt

number_of_games=10000
results=[game() for _ in range(number_of_games)]
steps=[result[0] for result in results]
sums=[result[1] for result in results]

#media lui S 

ioana=sum(sums) / number_of_games
print(f"Media lui S din {number_of_games} jocuri: {ioana} $")

#distributia 

plt.hist(sums, bins=20,edgecolor='pink', color='pink',alpha=0.7)
plt.title('Distributia sumei castigate de primul jucator (S)')
plt.xlabel('Suma castigata')
plt.ylabel('Frecventa')
plt.show
```

Daca moneda este masluita, avem alta probabilitate pentru fiecare eveniment, respectiv 0.7 si 0.3. Vom simula pe rand pentru fiecare dintre acestea. 

```python
# p=0.3

heads=0.3
results_p3=[game(heads) for _ in range (number_of_games)]
sums_p3=[result[1] for result in results_p3]
ioana_p3=sum(sums_p3) / number_of_games


# p=0.7

heads=0.7
results_p7=[game(heads) for _ in range(number_of_games)]
sums_p7=[result[1] for result in results_p7]
ioana_p7=sum(sums_p7) / number_of_games

print(f"Media lui S pentru p=0.3: {ioana_p3} $")
print(f"Media lui S pentru p=0.7: {ioana_p7} $")


plt.hist(sums_p3, bins=20,edgecolor='pink',color='pink', alpha=0.7, label ='p=0.3')
plt.hist(sums_p7, bins=20, edgecolor='pink',alpha=0.7, label ='p=0.7')
plt.title('Distributia sumei castigate de primul jucator (S) in cazul monedei masluite')
plt.xlabel('Suma castigata')
plt.ylabel('Frecventa')
plt.legend()
plt.show()
```

# Exercitiul 3

Probabilitatile de preluare a unui client sunt 3/13, 6/13 si 4/13 deoarece suma ratelor de servire este 3+6+4=13, fiecare probabilitate reprezentand raportul dintre rata respectiva de servire si suma tuturor ratelor. 

```python
import numpy as np
import matplotlib.pyplot as plt
import arviz as az

# parametrii pt distributiile exponentiale

lambdas=[3,6,4]

probabilities=[l/sum(lambdas) for  l in lambdas]

number_of_simulations=10000   #clienti
service_times= np.zeros(number_of_simulations)

for i in range(number_of_simulations):
    frizer_index=np.random.choice(range(3),p=probabilities)
        
    service_times[i] =np.random.exponential(scale=1/lambdas[frizer_index])
    
    
avg_service_time=np.mean(service_times)
std_service_time=np.std(service_times)


print(f"Media timpului de servire: {avg_service_time:.4f} ore") 
print(f"Deviatia standard a timpului de servire: {std_service_time: .4f} ore")

az.plot_kde(service_times)
plt.title('Densitatea aproximativa a timpului de servire')
plt.xlabel('Timp de servire (ore)')
plt.ylabel('Densitate')
plt.show()
        

```
